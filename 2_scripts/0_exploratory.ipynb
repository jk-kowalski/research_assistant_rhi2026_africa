```python
import os
import pypandoc

def convert_docx_to_markdown(docx_path, markdown_path=None):
    """
    Convert a Word document (.docx) to Markdown using pypandoc.
    
    Args:
        docx_path (str): Path to the input .docx file
        markdown_path (str, optional): Path to the output .md file. 
                                      If None, uses the same name as input with .md extension
    
    Returns:
        str: Path to the generated Markdown file
    """
    # Ensure pandoc is installed
    pypandoc.ensure_pandoc_installed()
    
    if not os.path.exists(docx_path):
        raise FileNotFoundError(f"Input file not found: {docx_path}")
    
    if markdown_path is None:
        markdown_path = os.path.splitext(docx_path)[0] + '.md'
    
    # Use pypandoc to convert docx to markdown
    try:
        pypandoc.convert_file(docx_path, 'md', outputfile=markdown_path)
        
        print(f"Successfully converted {docx_path} to {markdown_path}")
        return markdown_path
        
    except Exception as e:
        print(f"Error converting file: {e}")
        raise

# Example usage:
# Set working directory to the project root
os.chdir(r"C:\Users\kubak\Documents\research_assistant_wur\research_assistant_rhi2026_africa")

convert_docx_to_markdown('rhi2026_africa_RA_documentation.docx', 'rhi2026_africa_RA_documentation.md')
# or
# convert_docx_to_markdown('input.docx')  # outputs to input.md
```
